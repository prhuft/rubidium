```python
from datetime import datetime as DT
import matplotlib.pyplot as plt
import numpy as np
```

```python
DT(2021,5,6,11,30).strftime('%m/%d %H:%M')
```

```python
datetimes = ['2021,5,6,1,7', '2021,5,6,2,3', '2021,5,6,3,45', '2021,5,6,5,42', '2021,5,6,8,32', '2021,5,7,9,19', '2021,5,7,11,43', '2021,5,7,1,34', '2021,5,7,5,49',
             '2021,5,8,3,12', '2021,5,8,4,47']
xpts = [DT(*np.array(d.split(','), int)).strftime('%m/%d %H:%M') for d in datetimes]

ypts = np.array([1.2e-6, 1.6e-6, 2.2e-6, 3.4e-6, 8.6e-7, 2.2e-7, 1.9e-7, 1.9e-7, 1.6e-7, 1.2e-7, 1e-7])
ypts2 = np.array([26,37,72,107,108,108,109,109,108,108,109])
```

```python
fig,ax = plt.subplots(sharex=True)
ln1 = ax.plot(xpts, np.log10(ypts),c='b',label='Pressure')
ax.set_xticklabels(xpts, Rotation=45)
ax2 = ax.twinx()
ln2 = ax2.plot(xpts,ypts2,c='r',label='Temperature')
ax.set_ylabel("Log(Pressure [hPa])")
ax.set_xlabel("Date, Time")

lns = ln1 + ln2
labels = [l.get_label() for l in lns]

ax.legend(lns, labels, loc='center right')
plt.savefig('ffpr_bake_1_20210511.png',bbox_inches='tight')
plt.show()
```
