```python
from numpy import *
import matplotlib.pyplot as plt
```

```python
# 960 nm Toptica, 2019-12-9
mA = array([1, 5, 10, 15, 20, 25, 30, 35])
mW = array([.0052, .014, .03, .051, .080, 1.48, 4.08, 6.4])
assert len(mA)==len(mW)

fig, ax = plt.subplots()
ax.plot(mA,mW,marker="o",c="purple")
ax.set_xlabel("Diode Current [mA]")
ax.set_ylabel("Output Power [mW]")
ax.set_title("960 nm Toptica: I vs P")
plt.show()
```
