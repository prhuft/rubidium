```python
import matplotlib.pyplot as plt
```

measuring sidebands from ixblue EOM NIR-MPX900-LN-05 for Jim Reardon. the measurement was done with a scanning FP and PDA10A.

```python
RF = [17, 13, 11, 9]; # dbm
V_sb = [2.44,1.28,.84,.6] # mV
V_car = [7.6,10.24,10.9,11.3] # mV
RP_fraction = [sb/car for sb,car in zip(V_sb,V_car)]
plt.scatter(RF,RP_fraction)
plt.xlabel("3.036 GHz microwave power (dBm)")
plt.ylabel("one sideband fraction")
```

```python
RF = [17, 13, 11, 9]; # dbm
V_sb = [2.44,1.28,.84,.6] # mV
V_car = [7.6,10.24,10.9,11.3] # mV
V0 = (50*0.001)**(0.5)
V_RF = [V0*10**(P/20) for P in RF]
RP_fraction = [sb/car for sb,car in zip(V_sb,V_car)]
plt.scatter(V_RF,RP_fraction)
plt.xlabel("3.036 GHz microwave voltage (V)")
plt.ylabel("one sideband fraction")
```
