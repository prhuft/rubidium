```python
import matplotlib.pyplot as plt
from numpy import *
import csv
import os
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
```

```python
def process_csv(fname, newline='', cols=True, skiprows=1):
    rows = []
    with open(fname, 'r', newline=newline) as f:
        reader = csv.reader(f, delimiter=',')

        while skiprows > 0:
            reader.__next__()
            skiprows -= 1
        
        for row in reader:
            try:
                rows.append([float(x) for x in row if x != ''])
            except ValueError:
                print(f"row is problematic: \n {row}")

    if cols:
        return array(rows).T
    
    return rows
```

```python
# sample the file contents, get time increment
files = ['ffpr_locked_20210607.csv','ffpr_unlocked_20210607.csv']
fname = files[0]
newline=''
with open(fname, 'r', newline=newline) as f:
    reader = csv.reader(f, delimiter=',')
    for i,row in enumerate(reader):
        if i == 1:
            t_incr = float(row[5])        
        if i < 10:
            print(row)
```

```python
data = [process_csv(f, skiprows=2) for f in files]
data
```

```python
fig,axes = plt.subplots(2,1,sharex=True,dpi=200)
tpts = arange(0,len(data[0][1])*t_incr,t_incr)*1e3

ax,ax1 = axes

# locked data
ax.plot(tpts,data[0][2]*1e3,label='Locked')

# unlocked
ax.plot(tpts,data[1][2]*1e3,label='Unlocked')
ax.set_ylabel('Reflection PD [mV]')
ax.set_title('FFPR Reflection Signal')

ax1.plot(tpts,data[0][0],label='Ramp, locked')
ax1.plot(tpts,data[1][0],label='Ramp, unlocked')
ax1.set_xlabel('Time [ms]')
ax1.set_ylabel("Ramp \n(to Piezo Driver) [V]")

ax.legend(loc='right')
plt.show()
fname = 'ffpr_locked_vs_unlocked_20210607.png'
plt.savefig(fname,bbox_inches='tight')
```

```python
### this is a really fucking complicated way to do this. see if there is a way to just
### pass kwargs for creating the ax object when plt.subplots is called
def prestoplot(x, y, legend=None, loc='best', axeslabels=None, colors=None, 
               pltargs=[], pltkwargs={}, figargs=[], figkwargs={}):
    
    fig, ax = plt.subplots(*figargs, **figkwargs)
    
    try:
        datalength = len(x)
        
    except TypeError:
        x = [x]
        y = [y]
        legend = [legend]
        colors = [colors]
        
    for x,y,l,c in zip(x,y,legend,colors):
        ax.plot(x,y,label=l,color=c)
    
    if axeslabels != None:
        assert len(axeslabels) == 2
        ax.set_xlabel()
        
    ax.legend(loc=loc)
```

```python
def plot_linewidths(csvlist, labels, title, xlim):
    assert len(csvlist) == len(labels), "you need a label for each line!"
    
    fig,ax = plt.subplots()
    ax.set_title(title)
    ax.set_xlabel("[kHz]")
    ax.set_ylabel("[dBm]")
    ax.set_xlim(xlim[0], xlim[1])
    
    for i,csv in enumerate(csvlist):
        xpts = csv.xpts2 # xpts 2 should be the averaged curve 
        ypts = csv.ypts2 
        
#         x0 = csv.xpts2[where(csv.ypts2 == max(csv.ypts2))[0]][0]
        
        lwidth = get_linewidth(xpts, ypts)
        label = labels[i] + f'{lwidth} kHz @ -3dB'
        ax.plot((xpts-x0)/(1e3),ypts, label=label)
    
    ax.legend()
    plt.show()
```

```python
plot_linewidths([sh_780_ul,sh_780_zoom],["unlocked, ","locked, "], "Moglabs 780 Self-heterodyne 2020.09.18",[-500,500])
```
