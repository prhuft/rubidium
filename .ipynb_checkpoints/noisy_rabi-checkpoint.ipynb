```python
from numpy import *
import numpy.random as rand
import numpy.linalg as la
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from simfuncs import *
import os
```

### solve Schrodinger eq without any fancy symbolic stuff

```python
## simple Rabi oscillation in two level atom

psi0 = array([1,0], complex) # population starts in |g>

delta = 10
omega = 2*pi

# atom+field Hamiltonian, hbar=1
Haf = array([[0,-omega/2],[-omega/2,0]])

numsteps = 100
tmin = 0
tmax = 3 
tpts = linspace(tmin,tmax,numsteps)

dpsi = lambda t,state: -1j*dot(Haf,state)

soln = solve_ivp(dpsi,[tmin,tmax],psi0, t_eval=tpts,vectorized=True)
```

```python
labels = ['|g>','|e>']
for i,y in enumerate(soln.y):
    plt.plot(soln.t,real(conjugate(y)*y),label=labels[i])
plt.legend(loc='best')
```

## Noise from random processes - background

#### Now for some math. Taken from Xiaoyu's prelim slides. 

#### 1. background
X(t1), X(t2) are random variables where X is pulled some distribution function which gives us a set {X(t)}. 

Their correlation function is Rx(t1,t2) = E[Xt1Xt2]

Wide-sense stationary (WSS) variables:
 * E[X(t)] does not depend on t
 * E[X_{t+\tau}X_t] only depends on \tau = t1 - t2
 
Convenient for working with power spectral density Sx(f):
\int_f1^f2 Sx(f) \equiv power for f \elem (f1, f2)

Fourier relation between power spectral density and the correlation function:

 * Sx(f) = \mathcal{F}[Rx(\tau)]
 * Rx(\tau) = \mathcal{F}^-1[Sx(\tau)]

Time-series reconstruction

Known Sx to get X(t)

discretize Sx, get amplitudes Ax = \sqrt(Sx(fi)), get X(t) by taking \mathcal{F}^-1[Sx] with random phases

#### laser noise

E(t) is a random process if we give it random phases \phi(t)

Calculate R_E(\tau) and S_E(\nu), where the latter is what is gotten by a self-heterodyne measurement

Can show: phase and frequency noise power spectral densities are related by S_{\phi}(f) = S_{\nu}(f)/f^2

## Rabi oscillations with RIN

```python
## import actual RIN spectra
(f1,rin1), lbls1 = soln_from_csv(os.path.join('csv_analysis','soln_rin_780_20201223.csv'))
```

```python
plt.plot(f1,real(rin1[1:]),label=lbls1[1])
# plt.plot(f,10*log10(x1[1:]/x1[0]), label=testlbls[1])
plt.legend()
```

```python
## build time domain intensity function
int1 = lambda t: sum([sqrt(rin1[0]*10**(sx/10))*cos(2*pi*f*t+2*pi*rand.rand()) 
                      for f,sx in zip(f1,rin1[1:])]) # Fourier series
```

```python
## time-domain intensity
fig,ax = plt.subplots()
# use enough time steps to resolve all frequency components
t_end = 0.001 # [s]
numpts = 4*int(t_end*max(f1))
t_steps = linspace(0,t_end,numpts)
int1_steps = real(array([int1(t)+rin1[0] for t in tsteps]))
int1_steps /= max(int1_steps)
ax.plot(tsteps, int1_steps)
ax.set_ylim((0,1))
```
