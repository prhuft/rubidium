```python
import matplotlib.pyplot as plt
from numpy import *
```

```python
# single pass eff
pin = array([.5, 1, 1.5, 2, 2.5, 3, 3.7, 4.2, 4.7, 5.2, 5.7, 6.2, 6.7, 7.2]);
pout = array([20.3, 80, 180, 323, 506, 730, 1100, 1400, 1835, 2150, 2650, 
    3060, 3650, 4300])*10**-6;

fit = lambda x: 0.000024771 - 0.000024230*x + 0.000084721*x**2
fitpts = [fit(x) for x in pin]
```

```python
fig,ax = plt.subplots()
ax.scatter(pin,pout,label='0.000024771 -0.000024230 x+0.000084721 x^2') # from mathematica
ax.plot(pin,fitpts)
ax.set_xlim((0,7.5))
ax.set_ylim((0,.005))
ax.set_xlabel('Input 1540 [W]')
ax.set_ylabel('Output 770 [W]')
ax.set_title('Single-pass 770 conversion')
ax.legend()
```
