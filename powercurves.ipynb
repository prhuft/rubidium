```python
## Raman diode, replaced 9/19/19. died a day later

mA = linspace(10,80,8)

# pre-death
mWgood = array([.018,.04,5.57,19.88,31.8,42.5,55,65]) # after grating
uAgood = array([.7,1.9,180,553,890,1224,1550,1883]) # monitor 

# as it dies
mWbad = array([.013,.025,.042,.100])
uAbad = array([.7,1.3,2.7,3])

plt.plot(mA,mWgood,label='pre-death')
plt.plot(mA[:4],mWbad,label='final moments')
plt.xlabel("LD Current [mA]")
plt.ylabel("LD power after grating [mW]")
plt.legend()
plt.show()
```

```python
plt.plot(mA,uAgood,label='pre-death')
plt.plot(mA[:4],uAbad,label='final moments')
plt.xlabel("LD Current [mA]")
plt.ylabel("Monitor PD current [uA]")
plt.legend()
plt.show()
```
